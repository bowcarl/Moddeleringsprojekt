## Eurles Metode

Eurlers Metode en numerisk metode der vi dropper grenseverdiene, men gjør heller $h$ så liten som overhodet mulig. Nedenfor ser vi Newtons koffesient:

$f'(x) = \frac{f(x+h)-f(x)}{h}$

Her finner vi både en funksjon $f(x)$ samt den deriverte til funksjonen $f'(x)$. Målet er å finne denne funksjonen, derfor regner vi videre og får:

$f'(x) * h + f(x) = f(x+h)$

Til slutt omformer vi utrykket $f(x + h)$ slik at den kommer aleine på en side. Dette gjør fordi det er det neste skrittet videre fra $f(x)$ med skrittlengde $h$, og vi bruker derfor ofte en intialbetingelse for $f(x0)$.  

$f(x+h) = f(x)+ f'(x) * h$

Her har vi utledet Eurlers metode. Differansialikninger beskriver endring over tid og vi kan derfor skrive om steglengden $h$ til å representere endringen i tiden altså $\Delta_t$, noe som vil si at vi kan skrive det som $tk = t0 + kh$ hvor k representer et naturlig tall. Man kan finne funksjonsverdiene altså $f(tk)$ til en ukjent funksjon ved hvert tidsintervall/steg $tk$ ut fra ens deriverte samt en steglengde definert som h som representerer endringen i tid.

Eurles metode er har mange bruksområder samt er den lett og benytte seg av og avgir ganske gode resultater for funksjoner over kort tid. Ved Eulers metode finner du en tilnærmet funksjon.

## Eksempel kode

```python
from pylab import *


k = 0.05
N = 1000
T0 = 1
t0 = 0
tid_slutt = 100

dt = (tid_slutt-t0)/(N-1)

t = zeros(N)
T = zeros(N)
Tder = zeros(N)

t[0] = t0
T[0] = T0

T_0 = 20
for i in range(N-1):
    Tder[i] = - k*(T[i] - T_0)
    T[i+1] = T[i] + Tder[i]*dt
    t[i+1] = t[i] + dt


plot (t ,T )
title("Eksempel på Eulers Metode")
xlabel("tid")
ylabel("Temperatur")
show ()
```

Jeg valgte oppgave 1 hvor jeg skal ta for meg hvordan en befolkning på 500 mennesker i et isolert område med tilgang på noe medisin og våpen utvikler seg over en 100 års periode under en zombie-apokalypse. Jeg valgte at zombiene i starten bare skulle være 20. Her har jeg utviklet programmet gradevis og lagt til egne variabler og faktorer.
```python
from numpy import *
from math import *
from pylab import * #plot #linspace
from random import * #randint
```


## Statiske Variabler

Statiske variabler er de faktorene som ikke endres over tid samt ikke er gjort tilfeldige. Disse bruker jeg om gjennomsnittlige estimasjoner som dødsrate blant befolkningen, fødselsrate, innvandring og startmengder som zombier, mennesker og rasjoner. Jeg har prøvd og gjøre estimasjonene så realistiske som overhodet mulig.

```python
z0 = 20 #Start antall zombier
m0 = 500 #Start Antall Mennesker
prosent_fødsler_i_året = 8/100 #Menneske fødsler per år i prosent
fødselsrate = (m0/2) * prosent_fødsler_i_året #per kvinne i løpet av levetid
dødsrate = 3/100 #per 1000 innbygger i Alexandria
innvandring = 75 #per år sett bort ifra andre populasjoner
Mat = 500 #enheter hvor en enhet er tilnærmet en Matmengde per pers per år
Mattap = m0 #Hvor mye Mat spises opp i året i enheter
Mat_skaffelse = m0/6 # Hvor mye mat mann skaffer per år
sykdom_dødlighet = 1/10 #Hvor mange som dør av pandemier/epedemier.
Antall_Leger = 5/100 #Antall prosent leger i populasjonen
```

## Resultater

Deretter har jeg valgt og sette opp noen verdier som jeg anvender for å sjekke resultater. Jeg har gjort det mulig for alle senarioer i løpet av selvvalgt tidsperiode på 100år, de tre resultatene er alle mennesker omgjøres til zombier, zombiene dør ut, eller at begge overlever 100 år med konflikt.

```python
vinn = 0 #Zombiene dør ut
tap = 0 #Menneskene dør ut
DØD = 0 #Menneskene er utslettet
zDØD = 0 #Zombiene er utslettet
```

## Lister

Jeg anvender Lister for og legge til og trekke fra befolkingensutvikling/populasjonsdynamikk over en gitt tid

```python
Populasjon = [] #Menneskenes utvikling
Mat_Liste = [] #Mengden mat som skaffes i units
Zombie_Liste = [] #Zombienes utvikling
Pandemi = [] #Eventuelle pendemier
```

## Hovedløkke

Her legger jeg til varierende variabler i en løkke, hvor jeg har satt sansynlighetene til det jeg trodde ville bli mest realistisk, samt har jeg benyttet de ulike faktorene til og påvikre hverandre eks vær og mat. I tillegg til at jeg legger variablene i listene feks til Mat-listen eller zombie-listen slik at grafene kan variere fra gang til gang, selvom en computer heller ikke kan genere random tall i utgangspunktet, synes jeg det fremdeles er morsomt og se ulike grafer hver gang. 

```python
for i in range(100):
    Zombie_Naturlig_Død = randint(0, 6)/100 #Zombier naturlig død er mellom 1-6 sjans per 100 zombie dør
    Pandemi_Sansynlighet = randint(1, 50) # 1/50 sjans for en pandemi hvert år
    forsterkninger_Hilltop = randint(1, 100) #1/100 sjans for forsterkninger fra Hilltop
    forsterkninger_Kingdom = randint(1, 10) #1/10 sjans for forsterkninger fra Kingdom
    Sansynlighet_Menneske_Dreper_Zombier = randint(0, 4)/100 # 0-4 zombier per 100 zombie blir drept i året
    Zombietransformasjons_rate = randint(0, 3)/50 # Transformasjonsrater er på 0-3 i året per 10 zombier som infiserer dem
    Zombie_drepe_rate = randint(0, 4)/10 #0-4 sjans i året per 10 inbygger at zombiene dreper menneske
    Leger_Redder = randint(0, 5) #Legene har en sansynlighet på å redde 0-5 mennesker i året
    vær = randint(-1, 15)#-1-15 konstant på hvordan været begrenser muligheter for matsanking osv. 
    
    #Legger til listene
    Mat_Liste.append(Mat) #Oppdaterer listen
    Zombie_Liste.append(z0) #Oppdaterer listen
    Populasjon.append(m0) #Oppdaterer listen
```

## Variabelendringer

Her begynner vi og definere hvordan de ulike variblene påvirker popualsjonen og skaper en populasjonsdynamikk, her ser vi blant annet på de statiske variablene og deres påvirkning. Deretter benytter jeg meg av en varierende variabel som gir Alexandria forsterkinger ved en viss sansynlighet som er nevnt tidligere. Jeg syntes det var best hvis en var veldig sansynlig, men med liten påvirkning altså få tropper og den andre mer usansynlig, men da selfølgelig med masse tropper, noe som får grafene til og variere en del.

```python
    m0 += fødselsrate #Benytter de statiske variablene, og her får vi med hvor mange som fødes per innbygger i året
    m0 += innvandring #Deretter har vi en annen statisk variabel, som indikerer hvor mange som klarer og finne Alexandra og bosette seg der.
    m0 -= int(m0*dødsrate) #Legger dermed til dødsraten og her kan vi bare se på helttall, og har derfor lagt til int foran
    m0 -= (m0 * Zombietransformasjons_rate + Zombie_drepe_rate * m0) #Deretter legger jeg til varibalen som beskriver zombienes påvirkning på utviklingen i menneske populasjonen
    Mat -= Mattap #Her finner vi ut hvor mange inbyggere vi har og hvor mye mat som kreves og deretter oppdater mat-kurven seg
    Mat += Mat_skaffelse * vær #Her legges det til mengde mat man finner og variablen avhenger av vær-forholdet som også er varierende
    if Mat < 0: #Her sørger jeg for at mat ikke kan være en negativ verdi
        m0 += Mat
        Mat = 0 
    if forsterkninger_Hilltop == 100: #Her bruker vi den variernde variablen som en sansynlighet 1/100 sjans per år for forsterkninger
        m0 += 200 #legges til 200 i befolkningen
    if  forsterkninger_Kingdom == 10: #Derimot her får vi en sansynlighet på 1/10 per år
        m0 += 50 #Legges til til 50 i befolkningen
```

## Pandemi sansynlighet

Her velger jeg og lage en sansynlighet på at det vil skje en pandemi hvert år på 1/50 sjans og vi får i gjennomsnitt 2-4 pandemier i løpet av 100 år noe som jeg syntes var passende, jeg velger og vise dem sånn ca ved og legge til 500 (side by side) med populasjonen slik at det vises når disse pandemiene intreffer og hvor hardt de rammes. Jeg har i tillegg laget en statisk antall leger variabel samt en varierende varibel for kurrering av zombie-infeksjonen.

```python

    if Pandemi_Sansynlighet == 50: #Her gjør jeg nesten det samme som ved forsterkninger, legger til en 1/50 sjans for at det vil skje en pandemi hvert år
        Pandemi.append(500) #Appender 500 som er lik sum som start befolkningen bare for å kunne skje når pandemien skjer og utslaget den gir
        Helbredes = Antall_Leger * m0 * sykdom_dødlighet * Leger_Redder #Her sjekker vi hvor mange som helbredes/kureres
        m0 -= m0 * sykdom_dødlighet #Oppdaterer igjen befolkningen med tanke generel sykdomsdødlighet
        m0 += Helbredes #Til slutt oppdateres befolkningen med hensyn på hvor mange som helbredes
    else:
        Pandemi.append(0) #Sørger bare for at det ikke skjer noen utslag
        
    if m0 <= 0 or DØD == 1: #Her sjekker vi om Menneskene er under 0 altså døde slik at det ikke går lenger
        m0 = 0 #Setter dermed befolningen til 0
        DØD = 1 #Da er vi døde, og bruker dette for og printe ut resultatet til slutt
        Mat = 0 #Når menneskene dør ut er det heller ikke vits og vise mat-grafen
        tap = 1 #Dette er for å indekere at zombiene har vunnet
        
```

## Zombienes Populasjonsutvikling

Her har jeg laget populajons utviklingen til zombiene og hvordan de både kan dø naturlig og av menneske, samt transformasjonsraten som oppdaterer antall zombier. Deretter har jeg en telle variabel som sjekker om zombiene er mindre eller lik 0, da skal den stoppe og vi får opp at menneskene har vunnet.

```python
    z0 += (Zombietransformasjons_rate * m0 - (m0 * Sansynlighet_Menneske_Dreper_Zombier)) #Her oppdaterer jeg antall zombier ved bruk av de ulike variablene
    z0 -= z0 * Zombie_Naturlig_Død #Det samme gjøres her når vi ser på fratrekket
    
    if z0 <= 0 or zDØD == 1: #Sjekker om zombiene populasjonen er under 0 og sjekker om menneskene har vunnet
        vinn = 1 #Fører til at menneskene vinner
        zDØD = 1 #Fører til at zombiene er døde
        z0 = 0 #Populasjonen til zombier forblir 0 etter død
```

## Linspace og Plotting

Her bruker jeg linspace for og lage x-verdier som jeg definerer som år og deretter plotter jeg populasjonen, samt de eventuelle pandemiene som er utslagsgivende ved 500 i y retning. Så var det Mat-lista som ofte bestemmes av befolkningens utvikling og de er veldig avhengig av hverandre, lite mat liten befolkning osv, men kan også være motsatt. Til slutt er det en graf som viser Zombienes populasjonsvekst hvor jeg har bestemmt at når menneskene dør ut vil også zombiene dø ut fordi de ikke har noen annen måte og formere seg på, de vil derfor dø naturlig over tid. Jeg har også sørget for at ingen av grafene går under eller lik 0 og resultatet printes til slutt som nevnt tidligere

```python
x = linspace(0, 100, 100) #lager x verdier som jeg definerer som år, velger å se på en 100år tidsperiode

plot(x, Populasjon) #Her plotter jeg befolkningen over tid i Alexandria
grid("on") #legger til et rutenett
title("Populasjondynamikk i Alexandria") #Legger til en tittel
xlabel("år") #Legger til et navn på x-aksen
ylabel("populasjon") #Legger til et navn på y-aksen
plot(x, (Pandemi)) #Plotter eventuelle pandemier i løper av 100år
grid("on") #setter også her på et rutenett
show() #Til slutt legger jeg til show for og sikre at begge plottene kommer i samme graf

plot(x, (Mat_Liste)) #Deretter plotter jeg matlisten likt som med alle andre grafer sett ifra et 100år perspektiv
grid("on") #Legger til et rutenett
title("Byens Matlager") #Legger til en tittlel
xlabel("Antall År") #Legger til et navn på x-aksen
ylabel("Mat Enheter")#Legger til et navn på y-aksen
show() #Til slutt legger jeg til show for og sikre at det blir en separrert graf

plot(x, (Zombie_Liste)) #Her plotter jeg zombienes utviklin over tid
grid("on") #legger til et rutenett
title("Zombienes utvikling") #Legger til en tittlel
xlabel("Antall År") #Legger til et navn på x-aksen
ylabel("Antall Zombier") #Legger til et navn på y-aksen
show() #Helt til slutt legger jeg til show for og sikre at det blir en separrert graf

#Printer Resultatet
if vinn == 1: #Sjekker om tellevariablen er 1 hvis den er 1 så har menneskene vunent
    print("Vi vant") #printer derfor vi vant for å indekere dette
if tap == 1: #Derimot hvis tellevariablen her blir 1 vil det indekre på at zombiene vant
    print("zombiene vant") #Printer derfor at zombiene vant for å indekere dette
if vinn == 0 and tap == 0: #Hvis ikke dette skjer så forsetter kampen videre
    print("Kampen fortsetter") #Printer derfor at kampen forsetter for å indekere dette

```

# Hele koden

```python
from pylab import *
from math import *
from random import *

#statiske variabler
z0 = 20 #Start antall zombier
m0 = 500 #Start Antall Mennesker
prosent_fødsler_i_året = 8/100 #Menneske fødsler per år i prosent
fødselsrate = (m0/2) * prosent_fødsler_i_året #per kvinne i løpet av levetid
dødsrate = 3/100 #per 1000 innbygger i Alexandria
innvandring = 75 #per år sett bort ifra andre populasjoner
Mat = 500 #enheter hvor en enhet er tilnærmet en Matmengde per pers per år
Mattap = m0 #Hvor mye Mat spises opp i året i enheter
Mat_skaffelse = m0/6 # Hvor mye mat mann skaffer per år
sykdom_dødlighet = 1/10 #Hvor mange som dør av pandemier/epedemier.
Antall_Leger = 5/100 #Antall prosent leger i populasjonen

#Resultater
vinn = 0
tap = 0
DØD = 0
zDØD = 0

#Lister
Populasjon = []
Mat_Liste = []
Zombie_Liste = []
Pandemi = []

#Hovedløkke
for i in range(100):
    #Varierende Variabler
    Zombie_Naturlig_Død = randint(0, 6)/100
    Pandemi_Sansynlighet = randint(1, 50)
    forsterkninger_Hilltop = randint(1, 100)
    forsterkninger_Kingdom = randint(1, 10)
    Sansynlighet_Menneske_Dreper_Zombier = randint(0, 4)/100
    Zombietransformasjons_rate = randint(0, 3)/50
    Zombie_drepe_rate = randint(0, 4)/10
    Leger_Redder = randint(0, 5) #hvert år
    vær = randint(-1, 15)#Vær variabel for 
    
    #Legger til listene
    Mat_Liste.append(Mat)
    Zombie_Liste.append(z0)
    Populasjon.append(m0)
    
    #Variabelendringer
    m0 += fødselsrate
    m0 += innvandring
    m0 -= int(m0*dødsrate)
    m0 -= (m0 * Zombietransformasjons_rate + Zombie_drepe_rate * m0)
    Mat -= Mattap
    Mat += Mat_skaffelse * vær
    if Mat < 0:
        m0 += Mat
        Mat = 0 
    if forsterkninger_Hilltop == 100:
        m0 += 200
    if  forsterkninger_Kingdom == 10:
        m0 += 50
      
    
    #Pandemi sjekk
    if Pandemi_Sansynlighet == 50:
        Pandemi.append(500) #bare for å se utslaget
        Helbredes = Antall_Leger * m0 * sykdom_dødlighet * Leger_Redder
        m0 -= m0 * sykdom_dødlighet
        m0 += Helbredes
    else:
        Pandemi.append(0)
        
    if m0 <= 0 or DØD == 1:
        m0 = 0
        DØD = 1
        Mat = 0
        tap = 1
        
    #Zombienes populasjonsutvikling   
    z0 += (Zombietransformasjons_rate * m0 - (m0 * Sansynlighet_Menneske_Dreper_Zombier))
    z0 -= z0 * Zombie_Naturlig_Død
    
    if z0 <= 0 or zDØD == 1:
        vinn = 1
        zDØD = 1 
        z0 = 0
    
    
x = linspace(0, 100, 100) #lager x verdier som jeg definerer som år

plot(x, Populasjon)
grid("on")
title("Populasjondynamikk i Alexandria")
xlabel("år")
ylabel("populasjon")
plot(x, (Pandemi))
grid("on")
show()

plot(x, (Mat_Liste))
grid("on")
title("Byens Matlager")
xlabel("Antall År")
ylabel("Mat Enheter")
show()

plot(x, (Zombie_Liste))
grid("on")
title("Zombienes utvikling")
xlabel("Antall År")
ylabel("Antall Zombier")
show()

#Printer Resultatet
if vinn == 1:
    print("Vi vant")
if tap == 1:
    print("zombiene vant")
if vinn==0 and tap == 0:
    print("Kampen fortsetter")
```
